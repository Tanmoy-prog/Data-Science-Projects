```python
import pandas as pd
df=pd.read_csv('train.csv')
d=df.columns
df1=pd.DataFrame(df)
df1=df1.reindex(columns=['PassengerId', 'Pclass', 'Name', 'Sex', 'Age', 'SibSp',
       'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked','Survived'])
df1
```

```python
df1.drop(['PassengerId', 'SibSp',
       'Parch', 'Ticket', 'Cabin', 'Embarked','Name'],axis=1,inplace=True)
```

```python
df1
```

```python
target=df1.Survived
inputs=df1.drop('Survived',axis='columns')
```

```python
inputs
```

```python
dummy=pd.get_dummies(inputs.Sex)
dummy
```

```python
inputs=pd.concat([inputs,dummy],axis='columns')
```

```python
inputs.drop('Sex',axis='columns',inplace=True)
inputs
```

```python
inputs.columns[inputs.isna().any()]
```

```python
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(inputs,target,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
inputs
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
import pandas as pd
s=pd.read_csv('spam.csv',encoding = "ISO-8859-1")
```

```python
spam=pd.DataFrame(s[['catagory','message']])
pd.get_dummies(spam.catagory)
```

```python
spam['spam']=spam['catagory'].apply(lambda a:1 if a == 'spam' else 0)
```

```python
spam
```

```python
x_train,x_test,y_train,y_test=train_test_split(spam.message,spam.spam,test_size=0.2)
x_train
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
v=CountVectorizer()
v.fit(x_train.values)
x_train_count=v.transform(x_train.values)
```

```python
x_train_count.toarray()[1:3]
```

```python
from sklearn.naive_bayes import MultinomialNB
mnb=MultinomialNB()
```

```python
mnb.fit(x_train_count,y_train)
```

```python
mnb.score(x_train_count,y_train)
```

```python
emails=[
    'Hey mohan, can we get together to watch football game tommorrow?',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss the reward!'
]
emails_count=v.transform(emails)
mnb.predict(emails_count)
```
